```python
from matplotlib import pyplot as plt
from matplotlib import style
plt.plot([1,2,3],[4,5,2])
plt.title("Test Chart")
plt.xlabel("X axix")
plt.ylabel("Y axis")
plt.show()
```

```python
x = [1,3,6,8]
y = [2,4,3,6]

plt.plot(x,y)
plt.title("Data Chart")
plt.xlabel("data1")
plt.ylabel("data2")
plt.show()
```

```python
style.use('ggplot')

x = [2,4,9]
y = [4,2,6]

x2 = [2,6,10]
y2 = [6,2,8]

plt.plot(x,y,'g',label='line one',linewidth=5)
plt.plot(x2,y2,'c',label='line two',linewidth=5)

plt.title('Epic info')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.legend()
plt.grid(True,color='k')
plt.show()
```

```python
x = [1,3,5,7,9]
y = [2,4,6,4,5]

x2 = [2,4,6,8,10]
y2 = [6,2,8,6,7]

plt.bar(x,y,label='ex1')
plt.bar(x2,y2,label='ex2')

plt.title('Epic info')
plt.xlabel('bar height')
plt.ylabel('bar number')

plt.legend()
plt.show()
```
